```python
ratings = pd.read_csv("C:\\Users\\EECS\\Downloads\\ml-latest-small\\ml-latest-small\\ratings.csv")
ratings
```

```python
dataFrame = pd.DataFrame()
dataFrame['number_of_rating'] = ratings.groupby('movieId')['rating'].count()
dataFrame
```

```python
matrixRatingTable = ratings.pivot_table(index = 'userId', columns = 'movieId', values = 'rating')
matrixRatingTable
```

```python
def movie_rec():
    movie = input("What is the movieId to the movie that you like?: ")
    movie = int(movie)
    movie_user_ratings = matrixRatingTable[movie]
    similar_to_movie = matrixRatingTable.corrwith(movie_user_ratings)
    corr_to_movies = pd.DataFrame(similar_to_movie, columns = ['correlation'])
    corr_to_movies['number_of_rating'] = (dataFrame['number_of_rating'])
    corr_to_movies['movie title'] = (movies['title'])
    corr_to_movies.drop(corr_to_movies[corr_to_movies['number_of_rating'] < 150].index, inplace=True)
    corr_to_movies.sort_values(by='correlation', ascending = False, inplace = True)
    #reindex(corr_to_movies['number_of_rating'].sort_values(0, ascending = False).index)
    print(corr_to_movies.head(6))
```

```python
movie_rec()
```
